```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

distplot/histplot
jointplot
pairplot
rugplot

```python
df = sns.load_dataset("tips")
df
```

```python
# hisplot
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20 , kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20 , kde=True)
```

```python
# joinplot
sns.jointplot(x='total_bill' , y='tip' , data=df , kind='scatter',color='pink')
```

```python
sns.jointplot(x='total_bill' , y='tip' , data=df , kind='kde',color='green')
```

```python
sns.jointplot(x='total_bill' , y='tip' , data=df , kind='hex',color='purple')
```

```python
sns.jointplot(x='total_bill' , y='tip' , data=df , kind='reg',color='red')
```

PAIR PLOT

```python
sns.pairplot(df , hue='sex')
```

```python
sns.pairplot(df , hue='size' ,palette='rainbow')
```

RUG PLOT

```python
sns.rugplot(df)
```
